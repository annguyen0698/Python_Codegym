```python

class Reservation:
    def __init__(self, refnum, name, phonenum_email, checkin, checkout):
        self.refnum = refnum
        self.name = name
        self.phonenum_email = phonenum_email
        self.checkin = checkin
        self.checkout = checkout

    def __str__(self):
        return (f"Reference Number:{self.refnum}, Name: {self.name}, Phone Number/Email: {self.phonenum_email}"
                f"Check-in date: {self.checkin}, Check-out date: {self.checkout}")
        
from tkinter import *
from tkinter import messagebox


import datetime
date = datetime.datetime.today()
monthdayyear = date.strftime("%m%d%Y")
combined_date_time = date.strftime("%m/%d/%Y %H:%M")

import random

number_list = []

root = Tk()
root.title('Hotel Reservation System')
root.geometry("700x500")

#window frame
frame = LabelFrame(root, padx=80, pady=50)
frame.pack(padx=10, pady=10)

reservation_list = []

#logic for button click that will open the new window 
def button_click(function):
    number = random.randint(1, 20)
    while True:
        refnum_check = str(monthdayyear) + str(number)
        if refnum_check not in number_list:
            number_list.append(refnum_check)
            refnum_entry = refnum_check
            break
        else:
            number = random.randint(1, 20)
    
    if function == "Create":
        window1 = Toplevel()
        window1.title("Create Reservation")

        frame1 = LabelFrame(window1, text="Reservation System", padx=80, pady=50)
        frame1.pack(padx=10, pady=10)

        def create_reservation():
            refnum_entry = number_list[-1]
            ref_num = refnum_entry
            name_entry = name.get()
            phonenum_email_entry = phonenum_email.get()
            check_in_date_entry = check_in_date.get()
            check_out_date_entry = check_out_date.get()


            if name_entry == "" or phonenum_email == "" or check_in_date_entry == "" or check_out_date_entry == "":
                messagebox.showerror("Create Reservation", "Please enter all blank space")

            else:
                new_reservation = Reservation(ref_num, name_entry, phonenum_email_entry, check_in_date_entry, check_out_date_entry)
                reservation_list.append(new_reservation)
                list_reservations()
                messagebox.showinfo("New Reservation", f"Added {new_reservation} sucessfully")
                window1.destroy()

                
        #label making: 
        lbl_refnum = Label(frame1, text="Reference Number:")
        ref_number = Label(frame1, text = refnum_entry)
        ref_number.grid(row=1, column=1)
        lbl_name = Label(frame1, text="Name")
        lbl_phonenum_email = Label(frame1, text="Phone No./Email:")
        lbl_checkin = Label(frame1, text="Check In Date:")
        lbl_checkin_info = Label(frame1, text="mm/dd/yyyy")
        lbl_checkout = Label(frame1, text="Check Out Date:")
        lbl_checkout_info = Label(frame1, text="mm/dd/yyyy")
        
        #form label
        lbl_form = Label(frame1, text="RESERVATION FORM", font=("Arial", 20))
        lbl_form.grid(row=0, column=0, columnspan=2)

        #grid lable:
        lbl_refnum.grid(row=1, column=0)
        lbl_name.grid(row=2, column=0)
        lbl_phonenum_email.grid(row=3, column=0)
        lbl_checkin.grid(row=4, column=0)
        lbl_checkin_info.grid(row=4, column=2)
        lbl_checkout.grid(row=5, column=0)
        lbl_checkout_info.grid(row=5, column=2)
        
        #entry making: 
        name = Entry(frame1,width = 50)
        name.grid(row=2, column=1)
        phonenum_email = Entry(frame1, width = 50)
        phonenum_email.grid(row=3, column=1)
        check_in_date = Entry(frame1, width = 50)
        check_in_date.grid(row=4, column=1)
        check_out_date = Entry(frame1, width = 50)
        check_out_date.grid(row=5, column=1)
    
        #button making:
        btn_create = Button(frame1, text="Create", command = create_reservation)
        btn_create.grid(row=6, column=1)
        
    elif function == "Update":
        window2 = Toplevel()
        window2.title("Update Reservation")
        
        frame2 = LabelFrame(window2, text="", padx=80, pady=50)
        frame2.pack(padx=10, pady=10)
        
        def update_reservation():
            info_update = ent_update_info.get()
            check = False
        
            if info_update == "" :
                messagebox.showerror("Update Reservation","Please enter Reference number!!!")

            else:

                for i in range(len(reservation_list)):
                    if reservation_list[i].refnum == info_update:
                        update_window2 = Tk()
                        update_window2.title("Update Reservation")
                        update_window2.geometry("600x400")

                        # make frame 2.1 on the left side
                        frame21 = LabelFrame(update_window2, text="", padx=20, pady=20)
                        frame21.pack(side="left", padx=10, pady=10)
                        
                        lbl_form = Label(frame21, text="RECENT RESERVATION", font=("Arial", 10))
                        lbl_form.grid(row=0, column=1)
                        
                        lbl_recent_info = Label(frame21, text = f"Reference Number: {reservation_list[i].refnum}\nName: {reservation_list[i].name}\nPhone Number/Email: {reservation_list[i].phonenum_email}\nCheck-in date: {reservation_list[i].checkin}\nCheck-out date: {reservation_list[i].checkout}")
                        lbl_recent_info.grid(row=1, column=1)

                        # make frame 2.2 on the right side
                        frame22 = LabelFrame(update_window2, text="", padx=20, pady=20)
                        frame22.pack(side="right", padx=10, pady=10)
                        
#                        label in 2.2.:
                        lbl_form_2 = Label(frame22, text="UPDATE RESERVATION", font=("Arial", 20))
                        lbl_form_2.grid(row=0, column=0, columnspan=2)
                        lbl_refnum = Label(frame22, text="Reference Number: ")
                        lbl_refnum.grid(row=1, column=0)
                        ref_number = Label(frame22, text = info_update)
                        ref_number.grid(row=1, column=1)
                        
                        lbl_name_update = Label(frame22, text="Name Update: ")
                        lbl_name_update.grid(row=2, column=0)
                        lbl_name_update = Entry(frame22, width = 20)
                        lbl_name_update.grid(row=2, column=1)
                        
                        lbl_phoneemail_update = Label(frame22, text="Phone No./Email Update:")
                        lbl_phoneemail_update.grid(row=3, column=0)
                        phonenum_email_update = Entry(frame22, width = 20)
                        phonenum_email_update.grid(row=3, column=1)  
                        
                        lbl_checkin_update = Label(frame22, text="Check In Date Update:")
                        lbl_checkin_update.grid(row=4, column=0)
                        checkin_recent_update = Entry(frame22, width = 20)
                        checkin_recent_update.grid(row=4, column=1)  
                        
                        lbl_checkout = Label(frame22, text="Check Out Date:")
                        lbl_checkout.grid(row=5, column=0)           
                        checkout_recent = Entry(frame22, width = 20)
                        checkout_recent.grid(row=5, column=1)  

                        
                        check = True
                        window2.destroy()

                        break

                    if check == False:
                        messagebox.showerror("Search Reservation", "Cannot find reservation")
                
        #search label
        lbl_update = Label(frame2, text="UPDATE RESERVATION", font=("Arial", 20))
        lbl_update.grid(row=0, column=0, columnspan=2)

        #lbl for search
        lbl_ref_update = Label(frame2, text="Reference Number:  ")
        lbl_ref_update.grid(row=1, column=0)

        #entry making: 
        ent_update_info = Entry(frame2, width = 50)
        ent_update_info.grid(row=1, column=1)

        #button making:
        btn_update = Button(frame2, text="Update", command = update_reservation)
        btn_update.grid(row=3, column=1)

   
    elif function == "Search":
        window3 = Toplevel()
        window3.title("Search Reservation")
        
        frame3 = LabelFrame(window3, text="", padx=80, pady=50)
        frame3.pack(padx=10, pady=10)
        check = False
        
        def search_reservation():
            search_reservation = ent_search_info.get()
            check = False
            
            if search_reservation == "" :
                messagebox.showerror("Search Reservation","Please enter Reference number!!!")

            else:
                for i in range(len(reservation_list)):
                    if reservation_list[i].refnum == search_reservation:
                        messagebox.showinfo("Search Reservation", f"Reference Number:{reservation_list[i].refnum}, Name: {reservation_list[i].name}, Phone Number/Email: {reservation_list[i].phonenum_email}"
                            f"Check-in date: {reservation_list[i].checkin}, Check-out date: {reservation_list[i].checkout}")
                        check = True
                        window3.destroy()
                        break

                if check == False: 
                    messagebox.showerror("Search Reservation", "Cannot find reservation")

        #search label
        lbl_search = Label(frame3, text="SEARCH RESERVATION", font=("Arial", 20))
        lbl_search.grid(row=0, column=0, columnspan=2)

        #lbl for search
        search_info = Label(frame3, text="Reference Number:  ")
        search_info.grid(row=1, column=0)

        #entry making: 
        ent_search_info = Entry(frame3, width = 50)
        ent_search_info.grid(row=1, column=1)

        #button making:
        btn_search = Button(frame3, text="Search", command = search_reservation)
        btn_search.grid(row=3, column=1)

    
    elif function == "Delete":
            window4 = Toplevel()
            window4.title("Delete Reservation")
            frame4 = LabelFrame(window4, text="", padx=80, pady=50)
            frame4.pack(padx=10, pady=10)

            def delete_reservation():
                delete_refnum = ent_delete.get()

                if delete_refnum =="":
                    messagebox.showerror("Delete Reservation","Please enter Reference number!!!")

                else:
                    for i in range(len(reservation_list)):
                        check = False
                        if reservation_list[i].refnum == delete_refnum:
                            del_update_window = Tk()
                            del_update_window.title("Delete Booking")
                            del_update_window.geometry("600x200")
                            frame_del_update = LabelFrame(del_update_window, text="", padx=80, pady=50)
                            frame_del_update.pack(padx=10, pady=10)
                            check = False
                            def accept_delete(answer): 
                                if answer == "Yes":                
                                    for i in range(len(reservation_list)):
                                        if reservation_list[i].refnum == delete_refnum:
                                            del reservation_list[i]
                                            print([reservation.refnum for reservation in reservation_list])
                                            messagebox.showinfo("Delete Reservation", f"Delete Booking {delete_refnum} succesfully")
                                            break

                                elif answer == "No":
                                    del_update_window.destroy()
                            check = True
                            break

                        if check == False: 
                            messagebox.showerror("Delete Reservation", "Cannot find reservation")        
                            

                    lbl_del_noti = Label(frame_del_update, text = f"Are you sure you want to delete ref No. {delete_refnum}?", font=("Arial", 10))
                    lbl_del_noti.grid(row=0, column=0, columnspan=2)
                    btn_del_yes = Button(frame_del_update, text = "Yes",padx=20, pady=10, command=lambda: accept_delete("Yes"))
                    btn_del_yes.grid(row=1, column=0)
                    btn_del_no = Button(frame_del_update,text = "No", padx=20, pady=10,  command=lambda: accept_delete("No"))
                    btn_del_no.grid(row=1, column=1)

            #delete label
            lbl_delete = Label(frame4, text="DELETE RESERVATION", font=("Arial", 20))
            lbl_delete.grid(row=0, column=0, columnspan=2)

            #lbl for search
            delete_info = Label(frame4, text="Reference Number:  ")
            delete_info.grid(row=1, column=0)

            #entry making: 
            ent_delete = Entry(frame4, width = 50)
            ent_delete.grid(row=1, column=1)

            #button making:
            btn_delete = Button(frame4, text="Delete", command = delete_reservation)
            btn_delete.grid(row=3, column=1)

def list_reservations():
    print('*' * 20, end="\n")
    print('Reservation List:')
    for reservation in reservation_list:
        print('Reference Number:\t', reservation.refnum)
        print('Name:\t\t\t', reservation.name)
        print('Phone No./Email:\t', reservation.phonenum_email)
        print('Check In Date:\t\t', reservation.checkin)
        print('Check Out Date:\t\t', reservation.checkout)
        print(f'Reservation was made on:', combined_date_time)
        
def ref_check(refnum):
    for booking in reservation_list:
        if booking.refnum == refnum:
            return True
    return False

#welcome window buttons and labl
lbl_Welcome = Label(frame, text="WELCOME", font=("Arial", 20))
buttons = [
    Button(frame, text="CREATE RESERVATION", padx=50, pady=20, command=lambda: button_click("Create"), width=25),
    Button(frame, text="UPDATE RESERVATION", padx=50, pady=20, command=lambda: button_click("Update"), width=25),
    Button(frame, text="SEARCH", padx=50, pady=20, command=lambda: button_click("Search"), width=25),
    Button(frame, text="DELETE", padx=50, pady=20, command=lambda: button_click("Delete"), width=25),
    Button(frame, text="INVOICE", padx=50, pady=20, command=lambda: button_click("Invoice"), width=25),
]
#pack welcome window 
lbl_Welcome.pack()
for button in buttons:
    button.pack(ipadx=100)


root.mainloop()
```
