### V d: In ra tt c s chn t 0 n n

```python
n = int(input("Please input n: "))
i = 0  # khi to bin m
while i <= n:  # kim tra iu kin
    print(i)
    i += 2  # cp nht li gi tr bin m
```

### V d: In ra tt c s l t n v 1

```python
n = int(input("Please input n: "))
i = n
while i >= 1:
    if i % 2 == 1:
        print(i, end=" ")
    i -= 1
```

### V d: o ngc chui

```python
# "helloworld" -> "dlrowolleh"
content = input("Please input string: ")
i = len(content) - 1
while i >= 0:
    print(content[i], end="")
    i -= 1
```

### V d: Tam gic vung

```python
number_rows = int(input("please input number of row: "))
row = 1
while row <= number_rows:
    print("*"*row)
    row += 1
```

```python
# Nhp mt s n. In ra 3 s l u tin trong khong t n n 1.
n = int(input("Please input n: "))
i = n
count = 0  # theo di s lng s l  in ra

while i >= 1:
    if i % 2:
        print(i)
        count += 1
        
    if count == 3:
        break
        
    i -= 1
```

## Lnh continue

```python
# In ra tt c nhng s nguyn t 1 n 10, tr cc s 3, 4, 6.
i = 0
while i < 10:
    i += 1
    if i == 3 or i == 4 or i == 6:
        continue
    print(i)
    
```
