## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
n = int(input("Please input n: "))
i = 0  # khởi tạo biến đếm
while i <= n:  # kiểm tra điều kiện
    print(i)
    i += 2  # cập nhật lại giá trị biến đếm
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n = int(input("Please input n: "))
i = n
while i >= 1:
    if i % 2 == 1:
        print(i, end=" ")
    i -= 1
```

### Ví dụ: Đảo ngược chuỗi

```python
# "helloworld" -> "dlrowolleh"
content = input("Please input string: ")
i = len(content) - 1
while i >= 0:
    print(content[i], end="")
    i -= 1
```

### Ví dụ: Tam giác vuông

```python
number_rows = int(input("please input number of row: "))
row = 1
while row <= number_rows:
    print("*"*row)
    row += 1
```

### Ví dụ: 
Nhập vào một số nguyên a

Cho S(k) = 1 + 2 + 3 + ... + k

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
a = int(input("Please input a: "))
S = 0
k = 0

while S <= a:
    k += 1
    S += k
    
    
print(S)
print(k)
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
while True:
    print("Hello Python")
    break
```

```python
# Nhập một số n. In ra 3 số lẻ đầu tiên trong khoảng từ n đến 1.
n = int(input("Please input n: "))
i = n
count = 0  # theo dõi số lượng số lẻ đã in ra

while i >= 1:
    if i % 2:
        print(i)
        count += 1
        
    if count == 3:
        break
        
    i -= 1
```

## Lệnh continue

```python
# In ra tất cả những số nguyên từ 1 đến 10, trừ các số 3, 4, 6.
i = 0
while i < 10:
    i += 1
    if i == 3 or i == 4 or i == 6:
        continue
    print(i)
    
```
