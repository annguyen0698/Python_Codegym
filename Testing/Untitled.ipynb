```python
import tkinter as tk
from tkinter import ttk

# Define the class for the object that will be managed by the application
class Calculation:
    def __init__(self, first_number, second_number, operation):
        self.first_number = first_number
        self.second_number = second_number
        self.operation = operation

# Define the functions for the CRUD operations
def create_calculation():
    # Get the data from the user
    first_number = entry_first_number.get()
    second_number = entry_second_number.get()
    operation = entry_operation.get()

    # Create a new calculation object
    calculation = Calculation(first_number, second_number, operation)

    # Add the calculation object to the list of calculations
    calculations.append(calculation)

    # Update the TreeView
    treeview.insert("", "end", values=(calculation.first_number, calculation.second_number, calculation.operation))

def read_calculations():
    # Clear the TreeView
    treeview.delete(*treeview.get_children())

    # Add all the calculations to the TreeView
    for calculation in calculations:
        treeview.insert("", "end", values=(calculation.first_number, calculation.second_number, calculation.operation))

def update_calculation():
    # Get the data from the user
    index = treeview.selection()[0]
    first_number = entry_first_number.get()
    second_number = entry_second_number.get()
    operation = entry_operation.get()

    # Update the calculation object
    calculations[index].first_number = first_number
    calculations[index].second_number = second_number
    calculations[index].operation = operation

    # Update the TreeView
    treeview.set(index, values=(calculation.first_number, calculation.second_number, calculation.operation))

def delete_calculation():
    # Get the index of the calculation to be deleted
    index = treeview.selection()[0]

    # Delete the calculation object from the list of calculations
    calculations.pop(index)

    # Delete the calculation from the TreeView
    treeview.delete(index)

# Create the main window
window = tk.Tk()
window.title("Calculator Management System")

# Create the TreeView
treeview = ttk.Treeview(window)
treeview.pack()

# Create the labels and entry widgets
label_first_number = ttk.Label(window, text="First Number")
entry_first_number = ttk.Entry(window)
label_second_number = ttk.Label(window, text="Second Number")
entry_second_number = ttk.Entry(window)
label_operation = ttk.Label(window, text="Operation")
entry_operation = ttk.Entry(window)

# Pack the labels and entry widgets
label_first_number.pack()
entry_first_number.pack()
label_second_number.pack()
entry_second_number.pack()
label_operation.pack()
entry_operation.pack()

# Create the buttons
button_create = ttk.Button(window, text="Create", command=create_calculation)
button_read = ttk.Button(window, text="Read", command=read_calculations)
button_update = ttk.Button(window, text="Update", command=update_calculation)
button_delete = ttk.Button(window, text="Delete", command=delete_calculation)

# Pack the buttons
button_create.pack()
button_read.pack()
button_update.pack()
button_delete.pack()

# Define the list of calculations
calculations = []

# Run the application
create_calculation()
read_calculations()
```
