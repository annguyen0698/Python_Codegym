### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng

````python
kilograms = int(input("Input the number of kilograms need to be delivered: "))

if kilograms <= 10:
  delivery_cost = kilograms * 5
elif kilograms > 10 and kilograms <= 20:
  delivery_cost = 10 * 5 + (kilograms - 10) * 7
else:
  delivery_cost = 10 * 12 + (kilograms - 20) * 10

print("The delivery cost is $", delivery_cost)
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

````python
height = float(input("Enter your height in centimeters: "))
weight = float(input("Enter your weight in kilograms: "))

bmi = round(weight / (height / 100) ** 2,2)

if bmi < 16:
  classification = "Extremely thin (Level III)"
elif bmi < 17:
  classification = "Very thin (Level II)"
elif bmi < 18.5:
  classification = "Underweight (Level I)"
elif bmi < 25:
  classification = "Normal"
elif bmi < 30:
  classification = "Overweight"
elif bmi < 35:
  classification = "Obese Level I"
elif bmi < 40:
  classification = "Obese Level II"
else:
  classification = "Obese Level III"

print("Your BMI is:", bmi)
print("Your classification is:", classification)
````

### Bài 16: Viết một chương trình thực hiện phép tính đơn giản như sau:
```

    Kết quả: 16 + 9 = 25
```

````python
num1 = int(input("Input the first number: "))
num2 = int(input("Input the second number: "))
operator = input("Enter operators (+, -, *, /): ")

if operator == "+":
  result = num1 + num2
elif operator == "-":
  result = num1 - num2
elif operator == "*":
  result = num1 * num2
elif operator == "/":
  result = num1 / num2

print(f"Result: {num1} {operator} {num2} = {result}")
````
