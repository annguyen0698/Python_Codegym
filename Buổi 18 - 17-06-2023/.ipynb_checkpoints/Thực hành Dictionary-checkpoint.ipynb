## Bài 1: list to dict
Cho 2 list có cùng số phần tử với nhau. Hãy tạo ra một dict từ 2 list đã cho theo format dưới đây:

**Input**

```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
```

**Output**

```python
{'Ten': 10, 'Twenty': 20, 'Thirty': 30}
```

**Your code**

```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
# code below this line
dictionary = {}

for key in range(len(keys)):
    for value in range(len(values)):
        if key == value:
            dictionary[keys[key]] = values[value]
print(dictionary)
# code above this line
```

## Bài 7: key of minimum value
Hãy in ra key chứa value nhỏ nhất trong dictionary dưới đây.

**Input**

```python
sampleDict = {
  'Physics': 82,
  'Math': 65,
  'history': 75
}
```

**Output**

```python
Math
```

**Your code**

```python
sampleDict = {
  'Physics': 82,
  'Math': 65,
  'history': 75
}
# code below this line
vals = list(sampleDict.values())
min_val = vals[0]
for i in vals:
    if min_val>i:
        min_val = i
for k,v in sampleDict.items():
    if v = min_val:
        print(k)
# code above this line
```

## Bài 8: new salary
Hãy tăng lương(`salary`) cho Brad từ 6500 lên 8500 trong dictionary dưới đây nhé.

**Input**

```python
sampleDict = {
     'emp1': {'name': 'Jhon', 'salary': 7500},
     'emp2': {'name': 'Emma', 'salary': 8000},
     'emp3': {'name': 'Brad', 'salary': 6500}
}
```

**Output**

```python
sampleDict = {
     'emp1': {'name': 'Jhon', 'salary': 7500},
     'emp2': {'name': 'Emma', 'salary': 8000},
     'emp3': {'name': 'Brad', 'salary': 8500}
}
```

**Your code**

```python
sampleDict = {
     'emp1': {'name': 'Jhon', 'salary': 7500},
     'emp2': {'name': 'Emma', 'salary': 8000},
     'emp3': {'name': 'Brad', 'salary': 6500}
}
# code below this line

old_salary = sampleDict['emp3'].pop('salary')

print(f"Brad's old salary was {old_salary}")

sampleDict['emp3']['salary'] = 8500

print(sampleDict)
```
