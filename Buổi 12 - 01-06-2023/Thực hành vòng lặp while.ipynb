```python
'''
Bài 3: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
* - * - *
* - * -
* - *
* -
*
'''
```

```python
#cách 1: dùng while: 
number_rows = int(input("Please input the height of the triangle: "))

i = number_rows

while i > 0:
  j = 0
  while j < i:
    if j % 2 != 0:
      print('-', end=" ")    
    else:
      print('*', end=" ")
    j += 1
  print()

  i -= 1
```

```python
#cách 2: dùng for
number_rows = int(input("Please input the height of the triangle: "))

for i in range(number_rows, 0, -1):

    for j in range(0, i, 1):
    if j % 2 != 0:
      print('-', end=" ")    
    else:
      print('*', end=" ")
  print()
```

```python
#dùng while : 
n = int(input("input an integer: "))

i = 1
while i <= n:

    if i % 3 == 0:
        print("py")

    elif i % 7 == 0:
        print("thon")

    elif i % 3 == 0 and i % 4 == 0:
        print("python")

    else:
        print(i)

    i += 1
```

```python
'''
Bài 5: Viết chương trình nhập vào một số và kiểm tra xem số đó có phải là số nguyên tố hay không
Note: Số nguyên tố là số chỉ chia hết cho 1 và chính nó
'''
```

```python
#cách 1: dùng if và while
number = int(input("Enter a number: "))

if number > 1:

    i = 2
    while i <= int(number ** 0.5):
        if number % i == 0:
            print(number, "is not a prime number")
            break
        i += 1
    else:
        print(number, "is a prime number")

else:
    print(number, "is not a prime number")
```

```python
#cách 2: dùng if và for
number = int(input("Enter a number: "))

if number > 1:

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            print(number, "is not a prime number")
            break
    else:
        print(number, "is a prime number")

else:
    print(number, "is not a prime number")
```

```python
#cách 3: dùng def
def isPrime(num):
  for i in range(2, num):
    if (num % i) == 0:
      return False
  return True

num = int(input("Enter a number: "))

if num == 1:
  print(num, "is not a prime number.")
elif num > 1:
  if isPrime(num):
    print(num, "is a prime number.")
  else:
    print(num, "is NOT PRIME!")
```

```python
'''
Bài 6: Viết chương trình mô phỏng trò chơi kéo búa bao theo luật chơi như sau:
Quy ước: 1 - búa, 2 - kéo, 3 - bao
Khi trò chơi bắt đầu, yêu cầu người chơi nhập 1 số từ bàn phím đại diện cho kéo, búa, bao
Chương trình tự random ra một lựa chọn từ 1 đến 3.
Nếu người chơi thua thì tiếp tục chơi cho đến khi thắng thì kết thúc chương trình.
Ví dụ:

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 3
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 2
Máy tính chọn: 1
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 2
Người chơi thắng.
'''
```

```python
from random import randint

play_again = True

while play_again:
    player = str(input("You choose (Bao, Bua, Keo):").title())
    player = player.strip()

    computer = randint (0,2)
    if computer == 0:  
        computer = "Bao" 
    elif computer == 1:
        computer = "Bua"
    else :
        computer = "Keo"
    print("Computer chooses :"  + str(computer))

    if player == computer:
        print('draw')

    if player == "Keo":
            if computer == "Bao":
                print("You Win")
            else:
                print("You Lose")

    elif player == "Bao":
            if computer == "Bua":
                print("You Win")
            else:
                print("You Lose")

    elif player == "Bua":
            if computer == "Keo":
                print("You Win")
            else:
                print("You Lose")
        
    else:
            print('Wrong input')

    question = str(input("Would you want to play again? (y/n) ").title())
    if question == 'Y':
        play_again = True
    elif question == 'N':
        play_again = False
    

print("The game is finished, thanks for playing!")
```
