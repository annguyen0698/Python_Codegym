```python
import tkinter as tk
from tkinter import messagebox

class Bookstore:
    def __init__(self, title, genre, price):
        self.title = title
        self.genre = genre
        self.price = price

    def __str__(self):
        return f"Title: {self.title}, Genre: {self.genre}, Price: {self.price}"

# Create a window
window = tk.Tk()
window.title("Book Management System")
window.geometry("500x300")

book_list = []

def add_book():
    title_entry = title.get()
    genre_entry = genre.get()
    price_entry = price.get()

    if title_entry == "" or genre_entry == "" or  price_entry == "":
        messagebox.showerror("Add Book", "Please enter all three blank space")
    else:
        new_book = Bookstore(title_entry, genre_entry, int(price_entry))
        book_list.append(new_book)

    print([book.title for book in book_list])
    messagebox.showinfo("Book added", f"Added {new_book}")

def search_books():
    title_entry = title.get()

    if title_entry == "":
        messagebox.showerror("Search Books", "Please enter a book title")
    else:
        found_books = []
        for book in book_list:
            if book.title == title_entry:
                found_books.append(book)

        if found_books:
            for book in found_books:
                messagebox.showinfo("Search Book", f"Title: {book.title}, Genre: {book.genre}, Price: {book.price}")
        else:
            messagebox.showinfo("Search Books", "No books found")

def delete_books():
    title_entry = title.get()

    if title_entry == "":
        messagebox.showerror("Delete Books","Please enter a book title.")
        
    else:
        found_books = False
        for book in book_list:
            if book.title == title_entry:
                book_list.remove(book)
                found_books = True

        if found_books:
            print([book.title for book in book_list])
            messagebox.showinfo("Delete Books", f"Deleted {title_entry}.")
        else:
            messagebox.showinfo("Delete Books", "No books found.")

def update_price_books():
    title_entry = title.get()

    if title_entry == "":
        messagebox.showerror("Update Book Prices", "Please enter a book title.")
    else:
        found_books = False
        for book in book_list:
            if book.title == title_entry:
                new_price = price.get()
                book.price = int(new_price)
                found_books = True

        if found_books:
            print([book.title for book in book_list])
            messagebox.showinfo("Update Book Prices", f"Updated price of {title_entry} to {new_price}.")
        else:
            messagebox.showinfo("Update Book Prices", "No books found.")

title_label = Label(window, text="Title")
title = Entry(window)
genre_label = Label(window, text="Genre")
genre = Entry(window)
price_label = Label(window, text="Price")
price = Entry(window)

add_button = Button(window, text="Add Book", command=add_book)
search_button = Button(window, text="Search Books", command=search_books)
delete_button = Button(window, text="Delete Books", command=delete_books)
update_price_button = Button(window, text="Update Book Prices", command=update_price_books)

# Pack all the widgets
title_label.pack()
title.pack()
genre_label.pack()
genre.pack()
price_label.pack()
price.pack()
add_button.pack()
search_button.pack()
delete_button.pack()
update_price_button.pack()

window.mainloop()
```
